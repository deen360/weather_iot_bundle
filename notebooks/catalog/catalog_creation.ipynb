```python
from databricks.connect import DatabricksSession 
spark = DatabricksSession.builder.getOrCreate()
```

**CREATE CATALOGS AND VOLUME**

```python
catalog = dbutils.widgets.get("catalog")
schema_landing = dbutils.widgets.get("schema_landing")
schema_bronze = dbutils.widgets.get("schema_bronze")
schema_silver = dbutils.widgets.get("schema_silver")
schema_gold = dbutils.widgets.get("schema_gold")
volume = dbutils.widgets.get("volume")
```

```python
#CERATE THE CATALOG
spark.sql(f"""
CREATE CATALOG IF NOT EXISTS {catalog} COMMENT 'IoT Catalog for Weather Sensor Data'""")
```

```python
#CREATE Landing SCHEMA
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS {catalog}.{schema_landing}
COMMENT 'Landing zone for raw IoT sensor JSON files'""")
```

```python
#create volume to prevent to store checkpoints
spark.sql(f"""
CREATE VOLUME IF NOT EXISTS {catalog}.{schema_landing}.{volume}
COMMENT 'Landing zone for raw IoT sensor JSON files'
""")
```

```python
#CREATE Bronze SCHEMA
spark.sql(f""" CREATE SCHEMA IF NOT EXISTS {catalog}.{schema_bronze}
COMMENT 'Bronze zone for raw IoT sensor JSON files' """)
```

```python
#CREATE THE SILVER SCHEMA
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS {catalog}.{schema_silver}
COMMENT 'Silver zone for cleaned IoT sensor JSON files' """)
```

```python
#CREATE THE GOLD SCHEMA
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS {catalog}.{schema_gold}
COMMENT 'Gold zone for aggregated IoT sensor data' """)
```
